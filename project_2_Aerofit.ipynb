# Conditional Probabilities

Probability of each product given gender

```python
df1 = pd.crosstab(index=df['Gender'], columns=[df['Product']])
```

```python
df1
```

```python
p_781m = df1['KP781']['Male'] / df1.loc['Male'].sum()
p_481m= df1['KP481']['Male'] / df1.loc['Male'].sum()
p_281m = df1['KP281']['Male'] / df1.loc['Male'].sum()
```

```python
plt.bar(['KP781','KP481','KP281'],[p_781m,p_481m,p_281m])
plt.title('Probability of Product used by Male customer')
plt.xlabel('Procudts')
plt.ylabel('Probability')
```

Probability that a male customer buy Product KP281 is higher than he buys other two product

```python
p_781f = df1['KP781']['Female'] / df1.loc['Female'].sum()
p_481f= df1['KP481']['Female'] / df1.loc['Female'].sum()
p_281f = df1['KP281']['Female'] / df1.loc['Female'].sum()
```

```python
plt.bar(['KP781','KP481','KP281'],[p_781f,p_481f,p_281f])
plt.title('Probability of Product used by Female customer')
plt.xlabel('Procudts')
plt.ylabel('Probability')
```

It is clearly seen that female customer tend to buy product KP281 with greater probability then product KP481 and probability of buying product KP781 is least.

```python
df2 = pd.crosstab(index=df['MaritalStatus'], columns=[df['Product']])
```

```python
p_781p = df2['KP781']['Partnered'] / df2.loc['Partnered'].sum()
p_481p= df2['KP481']['Partnered'] / df2.loc['Partnered'].sum()
p_281p = df2['KP281']['Partnered'] / df2.loc['Partnered'].sum()
```

Partnered customer more likely to buy product KP281 than other two
